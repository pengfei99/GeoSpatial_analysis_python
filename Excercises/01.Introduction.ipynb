In this exercise, you'll investigate Kiva loans in the Philippines.  Can you identify regions that might be outside of Kiva's current network, in order to identify opportunities for recruiting new Field Partners?


```python
import geopandas as gpd
```

## Read Kiva loans data

Note the data have two important columns
- country
- geometry

```python
root_path="/home/pliu/data_set/kaggle/geospatial/L01"
loans_filepath = f"{root_path}/kiva_loans/kiva_loans.shp"

# Your code here: Load the data
world_loans = gpd.read_file(loans_filepath)
```

```python
world_loans.head()
```

## Read the geojson for country boundaries

Load a GeoDataFrame `world` containing country boundaries. Geopandas package provides a such file.


```python
# This dataset is provided in GeoPandas
world_filepath = gpd.datasets.get_path('naturalearth_lowres')
world = gpd.read_file(world_filepath)
world.head()
```

## Q1. Plot the loan on a map
Use the `world` and `world_loans` GeoDataFrames to visualize Kiva loan locations across the world.

The idea is very simple, we draw the first layer by using the `world` geo-dataframe (country boundaries). The second layer is the point of each loan.

```python
ax = world.plot(figsize=(10,10), color='none', edgecolor='gainsboro', zorder=3)
world_loans.plot(color='lightgreen', ax=ax)
```

## Q2 Select loans based in the Philippines.

Next, you'll focus on loans that are based in the Philippines. Create a GeoDataFrame `PHL_loans` which contains all rows from `world_loans` with loans that are based in the Philippines.

```python
PHL_loans =world_loans.loc[world_loans["country"]=="Philippines",:]
PHL_loans.head()
```

## Q3 Understand loans in the Philippines.

To plot the loans in the Philippines, we need a geo dataframe that describes the boundrays of different area of the Philippines. The `Philippines_AL258.kml` file contains boundaries for all islands in the Philippines. We will load it as a GeoDataFrame `PHL`.

```python
# Load a KML file containing island boundaries
gpd.io.file.fiona.drvsupport.supported_drivers['KML'] = 'rw'
PHL = gpd.read_file(f"{root_path}/Philippines_AL258.kml", driver='KML')
PHL.head()
```

```python
ax=PHL.plot(figsize=(10,10), color='none', edgecolor='gainsboro', zorder=3)
PHL_loans.plot(color='lightgreen', ax=ax)
```
